```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading Dataset
df = pd.read_csv('Ice Cream Ratings.csv')
df.set_index('Date', inplace = True)
df
```

```python
df.plot()
# df.plot(kind='line')
```

```python
df.plot(kind='line',subplots=True)
```

```python
df.plot(kind='line', title = 'Ice Cream Rating', xlabel = 'Daily Rating', ylabel = 'Scores')
```

```python
df.plot(kind='bar')
```

```python
df.plot(kind='bar', stacked = True)
```

```python
df['Flavor Rating'].plot(kind='bar', stacked = True)
```

```python
df.plot.barh(stacked = True)
```

```python
df.plot.bar(stacked = True)
```

```python
df.plot.scatter(x='Texture Rating', y= 'Overall Rating', s =100, c = 'Red')
#df.plot(kind='scatter',x='Texture Rating', y= 'Overall Rating')
```

```python
df.plot.hist(bins=20)  ## bins mean no. of distributions by default it is 10
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y='Flavor Rating', figsize=(10,6))
```
